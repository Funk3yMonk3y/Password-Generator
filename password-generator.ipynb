```python
import random
import string
import sys

print("\nGreetings Stranger :)!; Welcome to my random password-generator:")

a = int(input("\nPlease enter the desired length of your new password:"))

if a < 12 : print("\nDo yourself a favor and add more characters for your password, the point of this program is, to make sure your passwort is hard to guess") 
    

low = string.ascii_lowercase
caps = string.ascii_uppercase
num = string.digits
symb = string.punctuation

comb = low+caps+num+symb


if a >= 95: sys.exit
elif a<95: 
    mix = random.sample(comb,a)
    password = "".join(mix)
    print("\n"+password)

print("\nCya Mate, have a great day")
```
